```python
def trade_spider(max_pages):
    page=1
    
    level_0 = pd.DataFrame(columns  = ["Name", "URL"])
    
    while page <= max_pages:
        url = 'https://pureportal.coventry.ac.uk/en/organisations/coventry-university/persons/?page=' + str(page)
        source_code = requests.get(url)
        plain_text = source_code.text
        soup = BeautifulSoup(plain_text, "lxml")
        for link in soup.findAll('a', {'rel': 'Person'}):
            #href ="https://pureportal.coventry.ac.uk" + link.get('href')
            href ="" + link.get('href')
            title = link.string
            print(title)
            print(href)
            df2 = pd.DataFrame([[title,href]], columns= ["Name", "URL"])
            level_0 = level_0.append(df2)
            print("-----")
        page += 1
    #print(data)
    level_0.to_csv("level_0.csv", encoding="utf-8")

trade_spider(44)
```

```python
# Load data
level_0 = pd.read_csv("/content/drive/MyDrive/information retrieval/7071CEM Information Retrieval/CW/New folder/level_0.csv")
items_url = level_0["URL"]

def getProfileDetails_data(items_url):
    level_1 = pd.DataFrame(columns  = ["Name", "profile_URL", "publications_url"])

    for url in items_url:
        print(url)
        source_code = requests.get(url)
        plain_text = source_code.text
        soup = BeautifulSoup(plain_text, "lxml")

        #name of publisher
        for item_name in soup.findAll('header', {'class' : 'person-details'}):
            name = item_name.find('h1').text;
            print(name)

            #url
            publications_url = url + "/publications/"
            print(publications_url)
            print("----")
    
            df2 = pd.DataFrame([[name, url, publications_url]], columns= ["Name", "profile_URL", "publications_url"])
            level_1 = level_1.append(df2)

    level_1.to_csv("level_1.csv", encoding="utf-8")

getProfileDetails_data(items_url)
```
